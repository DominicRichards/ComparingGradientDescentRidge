```python
import numpy as np
from numpy.random import RandomState
import copy
import os
import matplotlib.pyplot as plt

import matplotlib.ticker as mticker
from scipy import stats
import math



#Make this notebook wider
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

import warnings
warnings.filterwarnings('ignore') 
```

```python
#Functions for computing the risks 
#Use notation alpha = \psi from the doc. alpha_plus is the upper bound on the singal strength, alpha_minus is a lowerbound. 


#The risk of a shinkage estimator. 
def FinalRisk(alpha_true,shrinkage):
    return( (1+alpha_true)* ( (shrinkage - 1/(1+alpha_true))**2))

#Optimal
def MinimaxRisk(k_discrete,alpha_minus,alpha_plus):
    return(((1/np.sqrt(1+alpha_minus) - 1./np.sqrt(1+alpha_plus))**2)/(k_discrete**2) )

#Construct the different collections of tuples
def OptimalList(k_discrete,alpha_minus,alpha_plus):
    x_plus = np.sqrt(1+alpha_plus)
    x_minus = np.sqrt(1+alpha_minus)
    c = (1./x_minus  - 1./x_plus)/float(k_discrete)
    l_list = np.zeros(k_discrete)
    for k in np.arange(k_discrete):
        l_list[k] = (1./x_plus + (k+1./2)*c)**2 - c*c/4. #Zero indexing for k, so j-1/2 for j=1,2,3, is the same as k+1/2 for k=0,1,2,3,..
    return(l_list)

def UniformList(k_discrete,alpha_minus,alpha_plus):
    return(np.linspace(0,1,k_discrete))

def GradientDescentList(k_discrete,alpha_minus,alpha_plus,Stepsize = 0.):
    return((1.-Stepsize)**(np.arange(k_discrete)))

def RidgeRegressionList(k_discrete,alpha_minus,alpha_plus):
    return(1-1./(np.linspace(0,1,k_discrete) +1))

#Given a collection of tuples, computes an estimate to the minimax error
def MinimaxError_Approx(List,alpha_minus,alpha_plus,Disc = 100):
    alpha_test = np.linspace(alpha_minus,alpha_plus,Disc)
    MaxError = 0
    for alpha_true in alpha_test:
        Tmp = np.min(FinalRisk(alpha_true,List))
        if(Tmp > MaxError):
            MaxError = copy.deepcopy(Tmp)
    return(MaxError)

#Theoretical values for the minimax risk  
def MinimaxError_Theory(EstimatorName,alpha_minus,alpha_plus,k_disc,Stepsize = 0):
    if(EstimatorName == "Optimal"):
        x_plus = np.sqrt(1+alpha_plus)
        x_minus = np.sqrt(1+alpha_minus)
        Output = ((1./x_minus - 1./x_plus)**2)/float(k_disc**2)
    if(EstimatorName == "Uniform"):
        b_plus = (k_disc - 1)/(1+alpha_plus)
        Output = 1./(4.*float(k_discrete-1)*(np.floor(b_plus) + 0.5))
    if(EstimatorName == "Gradient Descent"):
        b= 1-Stepsize
        h_x = np.log(1.+alpha_minus)/np.log(1./b)
        j_tmp = np.floor(h_x) 
        tau_minus = h_x - np.floor(h_x)
        
        tau_star = np.log(2./(1.+ b))/np.log(1./b)
        if(tau_star > tau_minus ):
            O_star_1 = (1 - b**tau_star)
        else:
            O_star_1 = np.min([ 1 - b**tau_minus,b*(b**(tau_minus- 1.) - 1.) ])
        
        if(tau_star < tau_minus):
            O_star_2 = (1. - b**tau_star)
        else:
            O_star_2 = np.min([ 1 - b**tau_minus,b*(b**(tau_minus- 1) - 1.) ])
        Output = (b**(j_tmp   - tau_star))* np.max([O_star_1**2,b*(O_star_2**2)])

    if(EstimatorName == "Ridge Regression"):
        phi = k_disc/alpha_minus -  (np.floor(k_disc/alpha_plus) +1)
        j_tmp = np.floor( k_disc/alpha_plus)
        IntPart =  k_disc/alpha_plus - np.floor(k_disc/alpha_plus)
        
        tau_star_1 = np.max( [ IntPart, (j_tmp + k_disc)/(2.*(k_disc + j_tmp) + 1.)] )
        tau_star_2 = np.min( [phi,(j_tmp + 1. + k_discrete)/(2*(j_tmp + 1. + k_discrete)  + 1.)])
        
        tmp_min_1 = np.min([tau_star_1 / (j_tmp + k_disc),(1-tau_star_1)/(j_tmp + 1+ k_disc)])
        tmp_min_2 = np.min([tau_star_2 / (j_tmp +1 + k_disc),(1-tau_star_2)/(j_tmp + 2 + k_disc)])
        
        O_star_1 = (k_disc/(j_tmp + tau_star_1))* tmp_min_1 /((1. + k_disc/(j_tmp + tau_star_1))**(1./2.))
        O_star_2 = (k_disc/(j_tmp + 1+ tau_star_2))* tmp_min_2 /((1. + k_disc/(j_tmp+1 + tau_star_2))**(1./2.))
        
        Output = np.max([O_star_1**2,O_star_2**2])
    
    return(Output)
```

```python
#Checking that the Theory matches practice i.e. 

Disc = int((10**(5))) #Used to estimate the minimax rate by discretising the interval [alpha_min,alpha_max]
k_test_no = 100 #Number of values for k to try 
k_disc_test = (10**(np.linspace(1.2,4,k_test_no))).astype("int") #Values of k to try
alpha_minus = 1.2
alpha_plus = 1000


MinimaxError_Estimated_stored = np.zeros((3,k_test_no))
MinimaxError_Theory_stored = np.zeros((3,k_test_no))
Labels = ["Gradient Descent","Optimal","Ridge Regression"]



# # ----------- Gradient Descent --------
for j,k_discrete in enumerate(k_disc_test):
    if(j % int(0.25 * k_test_no) == 0 ): print(str(j) + " / " + str(k_test_no) )
    Stepsize = 8./k_discrete
    GradientDescentList_tmp = GradientDescentList(k_discrete,alpha_minus,alpha_plus,Stepsize)
    MinimaxError_Estimated_stored[0,j] = MinimaxError_Approx(GradientDescentList_tmp,alpha_minus,alpha_plus,Disc)
    MinimaxError_Theory_stored[0,j] = MinimaxError_Theory("Gradient Descent",alpha_minus,alpha_plus,k_discrete,Stepsize)

    
# #----------- Optimal -----------------
for j,k_discrete in enumerate(k_disc_test):
    if(j % int(0.25 * k_test_no) == 0 ): print(str(j) + " / " + str(k_test_no) )
    OptimalList_tmp = OptimalList(k_discrete,alpha_minus,alpha_plus)
    MinimaxError_Estimated_stored[1,j] = MinimaxError_Approx(OptimalList_tmp,alpha_minus,alpha_plus,Disc)
    MinimaxError_Theory_stored[1,j] = MinimaxError_Theory("Optimal",alpha_minus,alpha_plus,k_discrete)

# # ----------- Ridge Regression -----------------
for j,k_discrete in enumerate(k_disc_test):
    if(j % int(0.25 * k_test_no) == 0 ): print(str(j) + " / " + str(k_test_no) )
    RidgeRegression_tmp = RidgeRegressionList(k_discrete,alpha_minus,alpha_plus)
    MinimaxError_Estimated_stored[2,j] = MinimaxError_Approx(RidgeRegression_tmp,alpha_minus,alpha_plus,Disc)
    MinimaxError_Theory_stored[2,j] = MinimaxError_Theory("Ridge Regression",alpha_minus,alpha_plus,k_discrete)


```

```python
np.savetxt("MinimaxError_Estimated.csv",MinimaxError_Estimated_stored,delimiter=",")
np.savetxt("MinimaxError_Theory.csv",MinimaxError_Theory_stored,delimiter=",")
```
