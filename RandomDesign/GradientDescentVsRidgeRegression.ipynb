```python
import numpy as np
from numpy.random import RandomState
import copy
import os
import matplotlib.pyplot as plt

import matplotlib.ticker as mticker
import math

#Make this notebook wider
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

import torch


import warnings
warnings.filterwarnings('ignore') 

```

```python
def SubOptimality_GD(Replications,Dimension,SampleSize,NoiseSD,Iterations,StepSize,r_squared,alpha,Lambda_min,device):
    EstError_GD = torch.zeros((Replications,Iterations),device=device)
    EstError_Ridge = torch.zeros((Replications,2),device = device)
    EstError_OptRidge = torch.zeros(Replications,device=device)
   
    Covar  = 1./(1 + torch.arange(Dimension,device=device))**(alpha)
   
    Opt_reg = NoiseSD*NoiseSD * Dimension /(r_squared*SampleSize)
   
    print("Opt Reg =  "+ str(NoiseSD*NoiseSD * Dimension /(r_squared*SampleSize)))
   
    for i in torch.arange(Replications):
        X = torch.normal(0, 1., size=(SampleSize,Dimension),device=device)*torch.sqrt(Covar)
        XTX = torch.matmul(torch.transpose(X,0,1),X)/float(SampleSize)
                             
        beta_star = (r_squared**(0.5))*torch.tensor([1],device=device).repeat(Dimension,1)/(Dimension**(0.5))
       
        Y = torch.matmul(X,beta_star) + torch.normal(0,1,size=(SampleSize,1),device=device)*NoiseSD
        XTY = torch.matmul(torch.transpose(X,0,1),Y)/float(SampleSize)
        beta_GD = torch.zeros((Dimension,1),device=device)
        for t in torch.arange(Iterations,device=device):
            beta_GD = beta_GD - StepSize * torch.matmul(XTX, beta_GD) + StepSize*XTY

            EstError_GD[i,t] = torch.sum( ( (beta_star-beta_GD)**2))
        beta_ridge = torch.linalg.solve( XTX + NoiseSD*NoiseSD*Dimension * torch.eye(Dimension,device=device)/(r_squared*SampleSize), XTY)
        EstError_OptRidge[i] = torch.sum( ( (beta_star-beta_ridge)**2))
        opt_ridge_pert_upper = Opt_reg + (1- Lambda_min )/(float(Iterations - 1))
        opt_ridge_pert_lower = Opt_reg - (1- Lambda_min )/(float(Iterations - 1))
       
        beta_ridge_pert  = torch.linalg.solve( XTX + opt_ridge_pert_upper* torch.eye(Dimension,device=device), XTY)
        EstError_Ridge[i,0] = torch.sum( ( (beta_star-beta_ridge_pert)**2))
        if(opt_ridge_pert_lower > 0 ):
            beta_ridge_pert  = torch.linalg.solve( XTX + opt_ridge_pert_lower* torch.eye(Dimension,device=device), XTY)
            EstError_Ridge[i,1] = torch.sum( ( (beta_star-beta_ridge_pert)**2))
        else:
            print("Negative Ridge!")
            EstError_Ridge[i,1] = EstError_Ridge[i,0]
       
        if(EstError_GD[i,-1] < EstError_GD[i,-2]):
            print("Not Converged!")
    return([EstError_GD,EstError_Ridge,EstError_OptRidge])

```

```python
# The experiments within the manuscript were run in batches for each value of alpha = 0.25, 0.1, 0.01


Replications = 20
Dimension = 5000
SampleSize = 200000
# NoiseSD  = 0.05
r_squared = 10.
alpha = 0.25
print("Min Eig. Value = " + str(1./(1+Dimension)**alpha))
device = "cuda:0"  #"cpu" to run on cpu 



globalReplications = 5
TestStore_all = []


NoiseTest = 10**(torch.linspace(1.3,0.8,10)) 

TmpMin = torch.zeros((len(NoiseTest),globalReplications),device=device)

OptRidgeReg = NoiseTest.min()*NoiseTest.min()*Dimension/(r_squared*SampleSize)
Lambda_min = 0.99*OptRidgeReg
StepSize = 1.
Iterations = int(1./(Lambda_min * StepSize))

for k,NoiseSD in enumerate(NoiseTest):
    TestStore = []
    print("Noise Iter =  " + str(k) + "/ 5")
    for i in torch.arange(globalReplications,device=device):
        print("Outer Replication : " + str(i+1) + " / " + str(globalReplications))
        Test1 = SubOptimality_GD(Replications,Dimension,SampleSize,NoiseSD,Iterations,StepSize,r_squared,Lambda_min,alpha,device)
        Tmpquant = (Test1[0].T - Test1[2]).mean(axis=1)/(Test1[1].T - Test1[2]).mean(axis=1).min()
        TestStore.append(Tmpquant)
        print("Minimum Ratio =" + str(Tmpquant.min()))
        TmpMin[k,i] = Tmpquant.min()
    TestStore_all.append(copy.deepcopy(TestStore))

```

```python
OptRidgeSequence  = Dimension*NoiseTest.cpu().numpy()**2/(r_squared*SampleSize)

plt.errorbar(OptRidgeSequence,y=TmpMin.cpu().numpy().mean(axis=1),yerr=2*TmpMin.cpu().numpy().std(axis=1),marker="o")
plt.axhline(1,color="grey")
plt.ylim([0.095,100])
plt.xlim([0.095,1.1])
plt.xscale("log")
plt.yscale("log")
plt.title("Subopt. Ratio Vs Noise Variance (" + r"$d = $"+ str(Dimension) + ", "  + r"$n = $" + str(SampleSize) + ")")
plt.ylabel("Sub Optimality Ratio " + r"$\mathcal{S}(\mathcal{C}^{GD},\mathcal{C}^{Ridge})$")
plt.xlabel("Optimal Regularisation " + r"$\lambda_\star$" )
plt.savefig("new_alpha_025/RatioVsOptReg1.pdf",dpi=128,bbox_inches="tight",pad_inches = 0)
plt.show()
```
